```python
def toy_function_1(x):
    y = x[0] * 4.0 + x[4] * (-0.5) + x[8] * 0.75  # Einzel Einflüsse root, no_deblock, no_weightb
    y = y + x[10] * 0.25 + x[11] * 0.35 + x[12] * 0.45  # Alternierende Gruppe lookahead
    y = y + x[18] * 0.2 + x[19] * 0.3 + x[21] * (-0.5)  # Alternierende Gruppe cores
    y = y + x[5] * x[8] * 0.5 + x[2] * x[5] * 0.75  # Interaction no_fast_pskip+no_weight, no_asm+no_fast_pskip
    #y = y + x[1] * x[2] * x[3] * 0.5 # Interaction 3x  'no_asm', 'no_8x8dct', 'no_cabac',
    return y
```

```python
for i in range(10):
    zeile = measurements.iloc[i, :22].to_list()
    print(zeile)
    toy_function_1(zeile)
```

```python
toy_werte = []
toy_werte_noise = []
for i in range(len(measurements)):
    zeile = measurements.iloc[i, :22].to_list()
    y = toy_function_1(zeile)
    toy_werte.append(y)
print(len(toy_werte))
```

```python
# Calculate 5% standard deviation based on the range of y_clean
y_range = np.max(toy_werte) - np.min(toy_werte)
noise_std = 0.05 * y_range

# Add noise to y values
toy_werte_noise = toy_werte + np.random.randn(len(toy_werte)) * noise_std
```

```python
print(toy_werte_noise)
print(toy_werte)
```

```python
measurements['toy_werte_2x'] = toy_werte
measurements['toy_werte_noise_2x'] = toy_werte_noise
```
